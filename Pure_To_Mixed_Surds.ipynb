```python
import math

def convert_pure_surd_to_mixed_surd(pure_surd,n):
    simplified_surd = ""
    
    # Check if the pure surd is a square root (√)
    if pure_surd.startswith("√"):
        number = int(pure_surd[1:])
        factors = prime_factors(number)
        simplified_surd = construct_mixed_surd(factors, "√",n)
    
    # Check if the pure surd is a cube root (∛)
    elif pure_surd.startswith("∛"):
        number = int(pure_surd[1:])
        factors = prime_factors(number)
        simplified_surd = construct_mixed_surd(factors, "∛",n)
    
    return simplified_surd

def prime_factors(number):
    factors = []
    while number % 2 == 0:
        factors.append(2)
        number //= 2
    for i in range(3, int(math.sqrt(number)) + 1, 2):
        while number % i == 0:
            factors.append(i)
            number //= i
    if number > 2:
        factors.append(number)
    return factors

def construct_mixed_surd(factors, given_format,n):
    mixed_surd = ""
    mixed_surd1=""
    count_dict = {}
    for factor in factors:
        count_dict[factor] = count_dict.get(factor, 0) + 1
    
    for factor, count in count_dict.items():
        if count >n:
            
            
            mul=(count/n)
            mixed_surd += str(factor*mul)  
            
            count=count-n
            if(count==1):
                 mixed_surd1 +=  str(factor)   
        elif(count==n):
            mixed_surd += str(factor) 
        elif(count==1):
            mixed_surd1 +=   str(factor)
    mixed_surd=mixed_surd+"*"+mixed_surd1+ given_format+str(n)+")" 
    return mixed_surd

# Example usage






def ConvertPureSurdsToMixedSurds(number):
    
    
    
   
    res1=number
  
   
    
    test1=res1.find("^")
    
   
    number_between=res1
    power=number_between.find("^")
    slash="/"
    number1_2=int(number_between[:power])
    number_between1=number_between[power+1:]
    slash_index=number_between1.find(slash)
        
    pow_number1=int(number_between1[slash_index+1:-1])
        #pow_number1=float(1/pow_number1)
        
    
    
    factors = prime_factors(number1_2)
    print("factors",factors)
    simplified_surd = construct_mixed_surd(factors,"^(1/",pow_number1)
   
    Result=simplified_surd
   
    return(Result)
    
  
 



y2="12^(1/2)"

y3="50^(1/2)"

y4="81^(1/3)"

y5="1280^(1/4)"











final=ConvertPureSurdsToMixedSurds(y4)
print(final)
```
